## Pelembutan citra (Image smoothing)

## Operasi pendeteksi (edge detection)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/orange.jpg')
 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gaussian = cv2.GaussianBlur(gray,(3,3),0)
 
#canny
img_canny = cv2.Canny(img,100,200)


#sobel
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely
 
#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)
 
img_prewitt=img_prewittx + img_prewitty
 
# laplacian 
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
plt.subplot(1,2,1),plt.imshow(img), plt.title('original')
plt.subplot(1,2,2),plt.imshow(img_canny), plt.title('canny')
plt.show()
```

```python
plt.subplot(1,2,1),plt.imshow(img_sobelx), plt.title('sobel x')
plt.subplot(1,2,2),plt.imshow(img_sobely), plt.title('sobel y')
plt.show()
```

```python
plt.subplot(1,2,1),plt.imshow(img_prewittx), plt.title('prewitt x')
plt.subplot(1,2,2),plt.imshow(img_prewitty), plt.title('prewitt y')
plt.show()
```

```python
plt.subplot(1,2,1),plt.imshow(img_prewitt), plt.title('prewitt x+y')
plt.subplot(1,2,2),plt.imshow(laplacian), plt.title('laplacian')
plt.show()
```

## Penajaman citra (image sharpening)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/lenna.jpg')

# construct a sharpening filter
sharpen_kernel = np.array((
            [0, -1, 0],
            [-1, 5, -1],
            [0, -1, 0]), dtype="int")

sharpen = cv2.filter2D(img, -1, sharpen_kernel)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(sharpen, cv2.COLOR_BGR2RGB)

plt.subplot(121),plt.imshow(img,cmap = 'gray'),plt.title('original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img2,cmap = 'gray'),plt.title('sharpen')
plt.xticks([]), plt.yticks([])
plt.show()
```
