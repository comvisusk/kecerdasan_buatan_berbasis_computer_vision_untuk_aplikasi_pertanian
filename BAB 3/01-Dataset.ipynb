# Mengecek Data

Membaca data menggunakan read_csv()

```python
url = 'https://raw.githubusercontent.com/suskay88/ww2_weather/main/ww2_weather.csv'
dataset = pd.read_csv(url)
```

# Plotting Data

Menampilkan plot for MinTemp dan MaxTemp

```python
dataset.plot(x='MinTemp', y='MaxTemp', style='o')  
plt.title('MinTemp vs MaxTemp') 
plt.xlabel('MinTemp') 
plt.ylabel('MaxTemp') 
plt.show()
```

Menampilkan Nilai Average of Max Temperature

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
seaborn.distplot(dataset['MaxTemp'])
```
