**Operator logika**

```python
# Contoh program operator logika
a = True
b = False

# Logika AND
c = a and b
print ("%r and %r = %r" % (a,b,c))

# Logika OR
c = a or b
print ("%r or %r = %r" % (a,b,c))

# Logika Not
c = not a
print ("not %r  = %r" % (a,c))
```

**Operator bitwise**

```python
a = 00111100
b = 00001101

# Operasi AND
c = a & b
print ("a & b = %s" % c)

# Operasi OR
c = a | b
print ("a | b = %s" % c)

# Operasi XOR
c = a ^ b
print ("a ^ b = %s" % c)

# Operasi Not
c = ~a
print ("~a = %s" % c)
```
