#**For loop**

```python
warna = ['Merah', 'Biru', 'Kuning', 'Biru']
for i in warna:
  print(i)
```

Penggunaan dan penulisan for-loop

```python
# penggunaan for-loop untuk print index

bahan_bahan = ('kol', 'cabai', 'kunyit', 'garam')
for i, bahan in enumerate(bahan_bahan):
  print(i, bahan)
```

```python
# penggunaan for-loop untuk sorted

angka = [2,1,4,3,5,8,6,7,9,10]
for n in sorted(angka):
  print(n)
```

```python
# penggunaan for-loop untuk penjumlahan dari list angka

angka = [10,7,9,6,8,4,5,1,2,3]
iterasi = len(angka)
total = 0
for i in range(iterasi):
  total = total + angka[i]
print(total)
```

```python
# penggunaan for-loop untuk mengganti isi dari element list

bahan_bahan = ['kol', 'cabai', 'kunyit', 'garam']
print("bahan_bahan sebelumnya:", bahan_bahan)
for bahan in bahan_bahan :
  index = bahan_bahan.index(bahan)
  print("bahan sebelumnya:", bahan_bahan[index])
  bahan="gula"
  bahan_bahan[index]=bahan
  print("bahan setelahnya:", bahan_bahan[index])
print("bahan_bahan sesudahnya:", bahan_bahan)
```

```python
# penulisan for-loop tanpa list comprehension

kubik = []
for n in range(10):
  kubik.append(n**3)
kubik
```

```python
# penulisan for-loop dengan list comprehension

kubik = [n**3 for n in range(10)]
kubik
```

#**While-loop**

Penggunaan dan penulisan while loop

```python
# while-loop dengan kondisi berhenti

count = 0
while (count < 9):
  print("The count is: ", count)
  count = count + 1
```

```python
# while-loop dengan kondisi tidak berhenti

its_raining = True
while its_raining:
  print("G-Incubation")
  # we'll jump back to the line with the word "while" from here
print("It's not raining anymore. ")
```

```python
# while-loop dengan kondisi if/else

while True:
  answer = input("Is it raining? (y=yes, n=no) ")
  if answer == 'y':
    print("It's raining!")
  elif answer == 'n':
    print("It's not raining anymore.")
    break     # end the loop
else:
  print("Enter y or n.") 
```
