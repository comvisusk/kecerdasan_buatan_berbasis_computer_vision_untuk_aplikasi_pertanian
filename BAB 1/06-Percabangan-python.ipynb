#**Struktur percabangan if**

```python
# if pada kondisi terpenuhi

jam = 7
if jam <= 7:
  print("yuk mandi")
print("sekarang saatnya pergi!")
```

```python
# if pada kondisi tidak terpenuhi

jam = 8
if jam <= 7:
  print("yuk mandi")
print("sekarang saatnya pergi!")
```

#**Struktur percabangan if/else**

```python
# if-else pada kondisi terpenuhi

jam = 6
if jam <= 7:
  print("yuk mandi")
else :
  print("cuci muka dulu")
print("sekarang saatnya pergi!")
```

```python
# if-else pada kondisi tidak terpenuhi

jam = 8
if jam <= 7:
  print("yuk mandi")
else :
  print("cuci muka dulu")
print("sekarang saatnya pergi!")
```

#**Struktur percabangan if/elif/else**

```python
jam = 6
makanan = False
if jam <= 7 and makanan == True :
  print("persiapan matang")
elif jam <= 7 and makanan == False :
  print("lapar di perjalanan")
elif jam > 7 and makanan == True :
  print("wah bau badan")
else :
  print("gawat!")
print("sekarang saatnya pergi!")
```
